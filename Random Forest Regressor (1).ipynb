```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data
```

```python
insurance_data.info()
```

```python
np.unique(insurance_data["sex"])
```

```python
np.unique(insurance_data["smoker"])
```

```python
np.unique(insurance_data["region"])
```

```python
category_features = ["sex" , "smoker" , "region"]
```

```python
one_hot = OneHotEncoder()
```

```python
transformer = ColumnTransformer([("one hot" , one_hot , category_features )] , remainder ="passthrough" )
```

```python
transformer
```

```python
transformer_data = transformer.fit_transform(insurance_data)
```

```python
transformer_data
```

```python
pd.DataFrame(transformer_data)
```

```python
dummies = pd.get_dummies(insurance_data , columns=["sex" , "smoker" , "region"])
```

```python
dummies
```

```python
x = dummies.drop("charges" , axis=1) 
y = dummies["charges"]
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.2) 
```

```python
model = RandomForestRegressor()
```

```python
model.get_params()
```

```python
model.fit( x_train , y_train )
```

```python
y_predict = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
np.random.seed()
```

```python
for i in range(10,151,10):
    print(f'trying model with {i} estimators')
    model = RandomForestRegressor(n_estimators = i).fit( x_train , y_train )
    print(f'model accuracy on test :{model.score( x_test , y_test )*100}')
    print(" ")
```
